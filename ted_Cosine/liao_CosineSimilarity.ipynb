```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
train=pd.read_csv("ted_main.csv")
train.head()
```

```python
print(pd.Series(train['tags']).describe(),'\n ')
```

```python
tags=train['tags'].tolist()
tfidf=text.TfidfVectorizer(input=tags)
matrix=tfidf.fit_transform(tags)
arr=tfidf.fit_transform(tags).toarray()
print(matrix.shape)
```

```python
arr
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_unigram=cosine_similarity(matrix)
```

```python
def get_similar_articles(x):
    return ",".join(train['title'].loc[x.argsort()[-6:-1]])
train['similar_articles_unigram']=[get_similar_articles(x) for x in sim_unigram]
```

```python
features = ['similar_articles_unigram','title','tags']
train[features]
```
