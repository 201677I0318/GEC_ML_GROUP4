```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
import nltk
transcripts = pd.read_csv("transcripts.csv")
ted_talks = pd.read_csv("ted_main.csv")
transcripts
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(use_idf=True,
                        ngram_range=(1,1), # considering only 1-grams
                        min_df = 0.005,     # cut words present in less than 0.5 of documents
                        max_df = 0.1)      # cut words present in more than 10% of documents 
tfidf_matrix = vectorizer.fit_transform(transcripts['transcript'])
print(tfidf_matrix.shape)
```

```python
### Get Similarity Scores using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
sim_unigram = cosine_similarity(tfidf_matrix)
```

```python
#create a column to record the vid number (index)
transcripts['index'] = transcripts.index.astype(str)
```

```python
def get_similar_articles(x):
    return ",".join(transcripts['index'].loc[x.argsort()[-4:-1]])
transcripts['similar_articles_unigram'] = [get_similar_articles(x) for x in sim_unigram]
transcripts
```
